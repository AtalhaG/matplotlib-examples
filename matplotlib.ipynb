```python
age_list=[10,20,30,30,60,89,90]
```

```python
weight_list=[89,67,78,40,20,20,40]
```

```python
np_array=np.linspace(0,10,20)
np_array2=np_array**2
np_array3=np_array**3
np_array4=np_array**4
np_array5=np_array**5
```

```python
plt.plot(np_array,np_array2)
```

```python
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(np_array,np_array2,"y")
plt.subplot(1,2,2)
plt.plot(np_array2,np_array3)
```

```python
plt.show()
```

```python
my_figure=plt.figure()
figure_axes=my_figure.add_axes([0.1,0.1,0.7,0.3])#,(x,x,yatay,dikey uzunluk)
figure_axes.plot(np_array,np_array2)
figure_axes.set_xlabel("x")
figure_axes.set_ylabel("y")
plt.show()
```

```python
new_fig=plt.figure(dpi=500)
new_Axes=new_fig.add_axes([0.1,0.1,0.9,0.9])
new_Axes.plot(np_array,np_array**2,label="np_array**2")
new_Axes.plot(np_array,np_array**3,label="np_array**3")
new_Axes.legend()
plt.show()
```
